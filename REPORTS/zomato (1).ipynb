##Step 1: Creating a Backup and Removing Irrelevant Columns

After successfully loading the original Zomato Bangalore Restaurants dataset, the next step is to prepare the data for cleaning and analysis. To ensure safety and reproducibility, I first created a backup copy of the original dataset. This protects the raw data from accidental modifications.

Next, I created a working copy that I will use for all further cleaning steps.
Since some columns—such as url, phone, and reviews_list—do not contribute meaningful value to the analysis, I identified and remove these irrelevant fields. This helps simplify the dataset, improves performance, and keeps the analysis focused on important attributes like ratings, cost, cuisine, and restaurant type.

This step ensures the dataset is cleanly structured and ready for detailed preprocessing.

```python

# STEP 1: Backup & Drop Irrelevant Columns


# Use the already loaded df_orig
print("Original Shape:", df_orig.shape)

# 1. Backup copy (do not modify)
df_backup = df_orig.copy()
print("Backup created (df_backup).")

# 2. Working copy
df = df_orig.copy()

# Columns that are not useful for analysis
drop_candidates = ['url', 'phone', 'reviews_list', 'menu_item', 'menu_items', 'menu']

# Drop only columns that actually exist
existing_to_drop = [col for col in drop_candidates if col in df.columns]

if existing_to_drop:
    print("Dropping irrelevant columns:", existing_to_drop)
    df.drop(columns=existing_to_drop, inplace=True)
else:
    print("No irrelevant columns found.")

print("\nNew Shape:", df.shape)

df.head()
```

##Step 2: Cleaning Rating and Cost Columns

In this step, I focus on cleaning two key numerical columns that are essential for meaningful analysis:

1. Cleaning the rate Column

The rate column contains several inconsistent entries such as:

"NEW"

"-"

"4.1/5"

These non-numeric values prevent proper analysis. To fix this:

I convert "NEW" and "-" into missing values (NaN)

Remove the /5 portion from ratings like "4.1/5"

Convert the entire column into a numeric data type

This ensures the rating values are clean and ready for numerical analysis.

2. Cleaning the Cost Column

The dataset contains a column named "approx_cost(for two people)", where values can appear as "1,200" or "250".
These cost values sometimes include commas, making them strings instead of numbers.

To clean this column:

I remove commas from cost values

Convert the cleaned values into numeric format

This prepares the cost column for comparisons, visualizations, and further analysis.

This cleaning step ensures that both rating and cost values are accurate, consistent, and numerically usable.

```python
import numpy as np


# 1. Cleaning the 'rate' column


if 'rate' in df.columns:
    print("Cleaning 'rate' column...")

    # Replace 'NEW' and '-' with NaN
    df['rate'] = df['rate'].replace(['NEW', '-', ' '], np.nan)

    # Remove '/5' if present
    df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False)

    # Convert to numeric
    df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

    print("Rate column cleaned.")
else:
    print("'rate' column not found.")



# 2. Cleaning the Cost Column


# Identify the correct cost column from the dataset
possible_cost_cols = ['approx_cost(for two people)', 'cost', 'approx_cost']
cost_col = None

for col in possible_cost_cols:
    if col in df.columns:
        cost_col = col
        break

if cost_col:
    print(f"Cleaning '{cost_col}' column...")

    # Remove commas
    df[cost_col] = df[cost_col].astype(str).str.replace(',', '')

    # Convert to numeric
    df[cost_col] = pd.to_numeric(df[cost_col], errors='coerce')

    print("Cost column cleaned.")
else:
    print("No cost column found.")


# Preview cleaned values
df[['rate', cost_col]].head()
```

##Step 3: Cleaning Categorical Columns, Handling Missing Values & Removing Duplicates

In this step, I focus on improving the consistency and quality of all non-numerical (categorical) data.
Many columns in the dataset—such as ‘location’, ‘rest_type’, ‘cuisines’, and ‘listed_in(type)’—contain inconsistencies such as:

Different capitalizations

Extra spaces

Mixed formatting

Missing values

To make the dataset clean and uniform, I perform the following actions:

1. Standardizing Categorical Columns

I convert all text values to lowercase and remove extra spaces.
This ensures that values like "Cafe", "cafe", and "Café " are treated consistently.

2. Handling Missing Values

Missing values can cause issues in grouping, filtering, and visualizations.
So, I:

Replace missing values in categorical columns with "unknown"

Keep missing values in numeric columns as NaN (because filling them may distort analysis)

3. Removing Duplicate Rows

Restaurant datasets often contain duplicate entries due to scraping from multiple sources.
Removing duplicates ensures:

Accurate statistics

Clean visualizations

Correct insights

This step finalizes the data cleaning process and prepares the dataset for feature engineering and EDA.

```python

# STEP 3: Clean Categorical Columns, Handle Missing Values, Remove Duplicates


import numpy as np


# 1. Standardize all categorical columns


cat_cols = df.select_dtypes(include='object').columns
print("Categorical Columns:", list(cat_cols))

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

print("\nCategorical columns cleaned.")


# 2. Handle Missing Values


# For categorical columns → fill with 'unknown'
for col in cat_cols:
    df[col] = df[col].replace(['nan', 'none', 'null'], np.nan)
    df[col] = df[col].fillna("unknown")

# For numeric columns → keep NaN (do not fill)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumeric Columns:", list(num_cols))

print("\nMissing values BEFORE cleaning:")
print(df.isnull().sum())

print("\nMissing values handled.")


# 3. Remove Duplicate Rows


before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]

print(f"\nRemoved {before - after} duplicate rows.")
print("Final Shape after cleaning:", df.shape)

# Preview cleaned data
df.head()
```

##Step 4: Creating Derived Columns (Feature Engineering)

To enhance the analytical power of the dataset, I create several new derived columns based on existing data. These engineered features help reveal deeper insights that are not immediately visible in the raw dataset.

1. Price Category

The cost column contains raw numerical values. To analyze affordability patterns, I grouped restaurants into:

Cheap

Moderate

Expensive

Premium

This classification makes cost-based analysis much more intuitive.

2. Rating Category

Numeric ratings are useful, but categorizing them makes customer satisfaction easier to interpret:

Poor

Average

Good

Excellent

3. Cuisine Count

Many restaurants serve multiple cuisines in one cell (e.g., “north indian, chinese, fast food”).
I count the number of cuisines to measure menu diversity.

4. Online Order & Table Booking Flags

I convert “yes/no” text values into:

1 (yes)

0 (no)

This makes statistical analysis easier and prepares the dataset for correlations and modeling.

These derived features prepare the dataset for deeper Exploratory Data Analysis (EDA).

```python

# STEP 4: Creating Derived Columns


import numpy as np

# Identify the correct cost column
possible_cost_cols = ['approx_cost(for two people)', 'cost', 'approx_cost']
cost_col = None

for col in possible_cost_cols:
    if col in df.columns:
        cost_col = col
        break


# 1. Price Category
def price_category(cost):
    if pd.isna(cost):
        return "unknown"
    if cost <= 300:
        return "cheap"
    elif cost <= 600:
        return "moderate"
    elif cost <= 1200:
        return "expensive"
    else:
        return "premium"

df['price_category'] = df[cost_col].apply(price_category)


# 2. Rating Category
def rating_category(rate):
    if pd.isna(rate):
        return "unrated"
    if rate < 2.5:
        return "poor"
    elif rate < 3.5:
        return "average"
    elif rate < 4.5:
        return "good"
    else:
        return "excellent"

df['rating_category'] = df['rate'].apply(rating_category)


# 3. Cuisine Count
df['cuisine_count'] = df['cuisines'].apply(lambda x: len(str(x).split(',')))


# 4. Online Order & Table Booking Flags
def yes_no_to_binary(x):
    x = str(x).strip().lower()
    if x == "yes":
        return 1
    elif x == "no":
        return 0
    else:
        return np.nan

df['online_order_flag'] = df['online_order'].apply(yes_no_to_binary)
df['book_table_flag'] = df['book_table'].apply(yes_no_to_binary)


# Preview new columns
df[['rate', 'rating_category', cost_col, 'price_category', 'cuisine_count', 'online_order_flag', 'book_table_flag']].head()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Use only first 3000 reviews to avoid memory crash
sample_reviews = df['reviews_list'].dropna().sample(3000, random_state=42)

# Join into text
review_text = " ".join(str(review) for review in sample_reviews)

wordcloud = WordCloud(
    stopwords=set(STOPWORDS),
    background_color='white',
    width=1200,
    height=600
).generate(review_text)

plt.figure(figsize=(14,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Customer Reviews", weight='bold')
plt.show()
```

I applied additional text cleaning and custom stopwords to remove encoding noise and generic words like “food”, “good”, “place”. This makes the word cloud focus on more specific aspects of customer experience such as taste, ambience, quantity and delivery.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import re

# Take a sample of reviews to avoid RAM issues
sample_reviews = df['reviews_list'].dropna().sample(3000, random_state=42)

# Basic cleaning function
def clean_review(text):
    text = str(text)
    text = text.replace('Rated', '').replace('/5', '')
    # remove those weird byte-like tokens and non-letters
    text = re.sub(r'Ã|Â|x..', ' ', text)
    text = re.sub(r'[^A-Za-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

cleaned_reviews = " ".join(clean_review(r) for r in sample_reviews)

extra_stops = set([
    'food', 'place', 'good', 'nice', 'restaurant', 'time',
    'service', 'really', 'one', 'great', 'also', 'ordered'
])

all_stopwords = STOPWORDS.union(extra_stops)

wordcloud = WordCloud(
    stopwords=all_stopwords,
    background_color='white',
    width=1200,
    height=600
).generate(cleaned_reviews)

plt.figure(figsize=(14,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Customer Reviews", weight='bold')
plt.show()
```
